```python
inputs = csv['Input'].unique()
```

```python
inputs
```

```python
df = pd.DataFrame()
meandata = pd.DataFrame()
for i in inputs:
    data = csv.loc[csv['Input'] == i]
    arraysizes = np.unique(data['ArraySize'])
    data_numpy = data.loc[data['Approach'] == 'numpy']
    data_cupy = data.loc[data['Approach'] == 'cupy']
    sd_numpy = np.zeros(len(arraysizes))
    y_numpy = np.zeros(len(arraysizes))
    sd_cupy = np.zeros(len(arraysizes))
    y_cupy = np.zeros(len(arraysizes))
    for i in range(0,len(arraysizes)):
        numpydata = data_numpy.loc[(data_numpy['ArraySize'] == arraysizes[i])]
        cupydata = data_cupy.loc[(data_cupy['ArraySize'] == arraysizes[i])]
        y_numpy[i] = np.mean(numpydata['Runtime(s)'])
        y_cupy[i] = np.mean(cupydata['Runtime(s)'])
        sd_numpy[i] = np.std(numpydata['Runtime(s)'])
        sd_cupy[i] = np.std(cupydata['Runtime(s)'])
        speedup = y_numpy[i]/y_cupy[i]
        df2 = pd.DataFrame({"Routine":["Relhum"],
                                    "Input":[data_numpy['Input'].unique()[0]],
                                    "ArraySize":[arraysizes[i]],
                                    "Speedup":[speedup],
                                    "SD_Numpy":[sd_numpy[i]],
                                    "SD_Cupy":[sd_cupy[i]]})
        meandata = pd.concat([meandata,df2])
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot(allData,name):
    arraysizes = np.unique(allData['ArraySize'])
    #y_numpy = allData.loc[(allData['Input'] == 'NumPy input')]
    #sd_numpy = y_numpy[["SD_Numpy", "SD_Numpy"]].max(axis=1)
    #y_Xarray = allData.loc[(allData['Input'] == 'Xarray with NumPy input')]
    #sd_Xarray = y_Xarray[["SD_Numpy", "SD_Numpy"]].max(axis=1)
    #y_Dask = allData.loc[(allData['Input'] == 'Xarray with Dask array input')]
    #sd_Dask = y_Dask[["SD_Numpy", "SD_Numpy"]].max(axis=1)
    # y_compute_xarray = allData.loc[(allData['Input'] == 'Compute Xarray with NumPy input ')]
    # sd_compute_xarray = y_compute_xarray[["SD_Numpy", "SD_Numpy"]].max(axis=1)
    y_compute_dask = allData.loc[(allData['Input'] == 'Xarray with Dask array input compute')]
    sd_compute_dask = y_compute_dask[["SD_Numpy", "SD_Numpy"]].max(axis=1)
    #y_numpy = y_numpy['Speedup']
    #y_Xarray = y_Xarray['Speedup']
    #y_Dask = y_Dask['Speedup']
    #y_compute_xarray = y_compute_xarray['Speedup']
    y_compute_dask = y_compute_dask['Speedup']
    plt.rcParams.update({'font.size': 22})
    fig, ax = plt.subplots(figsize=(12, 9))
    #ax.errorbar(arraysizes, y_numpy, yerr=sd_numpy, fmt='-o', label='Cupy over Numpy',markersize=20)
    #ax.errorbar(arraysizes, y_Dask, yerr=sd_Dask, fmt='-o', label='Xarray with Dask array',markersize=20)
    #ax.errorbar(arraysizes, y_Xarray, yerr=sd_Xarray, fmt='-.v', label='Xarray with NumPy',markersize=20)
    #ax.errorbar(arraysizes, y_compute_xarray, yerr=sd_compute_xarray, fmt=':+', label='Compute Xarray with NumPy input')
    ax.errorbar(arraysizes[0:5], y_compute_dask, yerr=sd_compute_dask, fmt='-s',  label='Compute Xarray with Dask array input')
    ax.legend();  # Add a legend.
    ax.set_xlabel('ArraySize')  # Add an x-label to the axes.
    ax.set_ylabel('Speedup (t_cpu/t_gpu)')  # Add a y-label to the axes.
    ax.set_title(("Relhum_Dask"))
    ax.set_xscale('log')
    #plt.savefig(name,dpi=fig.dpi)
plot(meandata,"Speedup_relhum_dask.jpg")
```

```python
import pandas as pd
csv = pd.read_csv("relhum_ported_test_xarray.csv")
#csv_compute = pd.read_csv("relhum_ported_compute_test.csv")
#csv_compute
import numpy as np
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot(allData,name):
    arraysizes = np.unique(allData['ArraySize'])
    sd_numpy = np.zeros(len(arraysizes))
    sd_cupy = np.zeros(len(arraysizes))
    y_numpy = np.zeros(len(arraysizes))
    y_cupy = np.zeros(len(arraysizes))
    plt.rcParams.update({'font.size': 30})
    for i in range(0,len(arraysizes)):
        cupydata = allData.loc[(allData['ArraySize'] == arraysizes[i]) & (allData['Approach'] == 'cupy')]
        numpydata = allData.loc[(allData['ArraySize'] == arraysizes[i]) & (allData['Approach'] == 'numpy')]
        y_cupy[i] = np.mean(cupydata['Runtime(s)'])
        y_numpy[i] = np.mean(numpydata['Runtime(s)'])
        sd_cupy[i] = np.std(cupydata['Runtime(s)'])
        sd_numpy[i] = np.std(numpydata['Runtime(s)'])
    fig, ax = plt.subplots(figsize=(12, 9))
    ax.errorbar(arraysizes, y_numpy, yerr=sd_numpy, fmt='-o',label='Xarray with Numpy',markersize=20)
    ax.errorbar(arraysizes, y_cupy, yerr=sd_cupy, fmt='-D',label='Xarray with Cupy',markersize=20)
    ax.legend();  # Add a legend.
    ax.set_xlabel('ArraySize')  # Add an x-label to the axes.
    ax.set_ylabel('Runtime(s)')  # Add a y-label to the axes.
    ax.set_title(("Relhum - Xarray including Numpy/CuPy array"))
    ax.set_xscale('log')
    #ax.set_yscale('log')
    #plt.savefig(name,dpi=fig.dpi)
```

```python
plot(allData,"Runtime_relhum_xarray_100chunk_changed_linear.jpg")
```
