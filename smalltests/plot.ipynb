```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('Test.csv')
```

```python
df
```

```python
#Test
x = df.loc[df['Approach'] == "NumPy"]['Array Size']
y_numpy = df.loc[df['Approach'] == "NumPy"]['Runtime']
y_cupy =df.loc[df['Approach'] == "CuPy"]['Runtime']
numpy_sd = np.asarray([3.375071461223291e-06,1.0987967130473053e-06,8.313939940530155e-07,1.2579828222535057e-06,
                       1.2520403619595573e-05,0.0001913389296351177,0.014182946665207186,0.021155865614997642,
                       0.122474415531762]),
cupy_sd = np.asarray([2.151482953415654e-06,5.876957523584423e-06,6.230940821105628e-06,5.209104954538285e-06, 
                      1.5652701760655996e-05, 8.649042676552321e-06,3.412342956415772e-05,0.0018119566344804313,
                      0.0016870744311483351])
```

```python
# Note that even in the OO-style, we use `.pyplot.figure` to create the Figure.
fig, ax = plt.subplots(figsize=(7, 4))
plt.xscale("log")
ax.errorbar(x, y_numpy, yerr=numpy_sd, fmt='-o')
ax.errorbar(x, y_cupy, yerr=cupy_sd, fmt='-o')
ax.plot(x, y_numpy, label='NumPy', color="steelblue")  # Plot some data on the axes.
ax.plot(x, y_cupy, label='CuPy',color="darkorange")  # Plot more data on the axes...
ax.set_xlabel('Array Size')  # Add an x-label to the axes.
ax.set_ylabel('Runtime (s)')  # Add a y-label to the axes.
ax.set_title("NumPy vs. CuPy")
ax.legend()
```
